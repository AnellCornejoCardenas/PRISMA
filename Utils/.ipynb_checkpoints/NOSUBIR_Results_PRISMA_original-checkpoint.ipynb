# Spectra information

```python
def spectra(out_path,ms):
    """
    This function allows us to visualize the spectra obtained from the spaxels of each galaxy.
    Inputs-> - out_path :  spectra directory
             - ms       :  stellar mass (as an array with the stellar mass of all the star into the spaxels)
    """
    files = !ls $out_path
    spec = [i for i in files if 'best_model' in i] #to get the modeled spectra files
    spec = [f"{out_path}/{i}" for i in spec] ##to get the full name of the spectra directory
    hdul = [fits.open(i) for i in spec] 
    wl = np.array([i[1].data['wavelength'] for i in hdul])
    L_norm_tot = np.array([i[1].data['L_lambda_total'] for i in hdul]) #data['stellar.young'] , data['nebular.continuum_young'] , data['nebular.lines_young']
    L_norm_young = np.array([i[1].data['stellar.young'] for i in hdul])
    L_norm_old = np.array([i[1].data['stellar.old'] for i in hdul]) 
    L_norm_nebular_lines_young = np.array([i[1].data['nebular.lines_young'] for i in hdul])
    L_tot = np.array([L_norm_tot[i] * ms[i] for i in range(len(ms))]) #L*M
    L_young = np.array([L_norm_young[i] * ms[i] for i in range(len(ms))]) #L*M
    L_old = np.array([L_norm_old[i] * ms[i] for i in range(len(ms))]) #L*M
    L_nebular_lines_young = np.array([L_norm_nebular_lines_young[i] * ms[i] for i in range(len(ms))]) #L*M
    
    wl_total = wl[0]
    L_total = np.sum(L_tot,axis=0)   
    L_total_young = np.sum(L_young,axis=0)   
    L_total_old = np.sum(L_old,axis=0)   
    L_total_nebular_lines_young = np.sum(L_nebular_lines_young,axis=0)   
    return(wl_total , L_total,L_total_young,L_total_old,L_total_nebular_lines_young)

def spectra_stars(out_path,ms):
    """
    This function allows us to visualize the spectra obtained from the spaxels of each galaxy.
    Inputs-> - out_path :  spectra directory
             - ms       :  stellar mass (as an array with the stellar mass of all the star into the spaxels)
    """
    files = !ls $out_path
    spec = [i for i in files if 'best_model' in i] #to get the modeled spectra files
    spec = [f"{out_path}/{i}" for i in spec] ##to get the full name of the spectra directory
    hdul = [fits.open(i) for i in spec] 
    wl = [i[1].data['wavelength'] for i in hdul]
    L_norm_tot = [i[1].data['L_lambda_total'] for i in hdul] 
    L_tot = [L_norm_tot[i] * ms[i] for i in range(len(ms))] #L*M
    
    L_total = np.sum(L_tot,axis=0)   
    return(wl , L_total)


def F_to_L(Flux , Dist=10,Watt=True):
    """
    Util function for transforming flux to luminosity considering a distance in pc.
    Input -> - Flux : flux in W/m^2
             - Dist : distance in pc 
    Output -> luminosity in Watts (if True) or erg/s
    """
    Dist = Dist* 30856775812799588 #meters 
    
    if Watt == True:
        return Flux * (4*np.pi * Dist * Dist)
    else: #erg/s
        L_W = Flux * (4*np.pi * Dist * Dist) * u.joule/u.s
        L_ergs = L_W.to(u.erg/u.s)#1W = 1e7 erg/s 
        return(L_ergs.value)


def lines_luminosities(line_name, model_path, m_gas, normalize = False):
    """
    With this function we can obtain the flux under the area of specific emission lines.
    Inputs ->   - line_name : name of the emission line (e.g., line.H-alpha,
                                                               line.H-beta,
                                                               line.OIII-500.7', ...)
                - model_path: path to the spectra directory
                - m_gas: mass of the gas clouds (array with the dimensions of the young populations)
                
    Output -> array with the line emission (in Watts(default) or erg/s) per young stellar population into the spaxel
               each spaxel.
    """
    models = fits.open(model_path)
    line_Fnu = models[1].data[line_name]  
    line_L_norm = F_to_L(line_Fnu, Watt = True)  #normalized to 1 M_sun, set Watt = False to get ergs/s
    if normalize == False:
        line_L = np.array([line_L_norm[i] * m_gas[i] for i in range(len(m_gas))]) #line*Mgas
    else: line_L = line_L_norm
    line_L = line_L[line_L != 0] #to remove the zero luminosity (equivalent to the older contribution)
    return line_L 
```

## Spaxel spectrum visualization

```python
wl_dic = {}
L_dic = {}
L_young_dic = {}
L_old_dic = {}
L_nebular_dic = {}

spx_ids = np.unique(intrin_data['Spax_id'])
cond = f"fesc{neb_f_esc_str}_fdust{neb_f_dust_str}"

for spx in spx_ids:
    
    ms = intrin_data[intrin_data["Spax_id"] == spx]['M_stars']   #stellar mass array
    
    spec_path = f"{spec_data_path}/out_{sim_name}-{galaxy_id}_{cond}_{int(spx)}" #spectra path 
    
    wl_spax , L_spax , L_young_spax , L_old_spax, L_nebular = spectra(spec_path,list(ms))
    #wl_spax , L_spax = spectra(spec_path,list(ms))
    wl_dic[spx] = wl_spax
    L_dic[spx] = L_spax
    L_young_dic[spx] = L_young_spax    
    L_old_dic[spx] = L_old_spax
    L_nebular_dic[spx] = L_nebular
```
